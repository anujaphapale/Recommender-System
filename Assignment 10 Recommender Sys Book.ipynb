```python
book1 = book.pivot_table(index='ID',
                columns='Book_Title',
                values='Book_Rating').reset_index(drop=True)
```

```python
book1
```

```python
#mapping of book ID with pivot table ID
book1.index=book.ID.unique()
```

```python
book1
```

```python
#Impute those NaNs with 0 values
book1.fillna(0, inplace=True)
```

```python
book1
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_sim = 1 - pairwise_distances( book1.values,metric='cosine')
```

```python
book_sim
```

```python
#Store the results in a dataframe
book_sim_df = pd.DataFrame(book_sim)
```

```python
book_sim_df
```

```python
#Set the index and column names to user ids 
book_sim_df.index = book.ID.unique()
book_sim_df.columns = book.ID.unique()
```
